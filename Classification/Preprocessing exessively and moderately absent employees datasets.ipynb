# Import the relevant libraries

```python
import numpy as np
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('C:\\Users\\HP\\Downloads\\Absenteeism_data.csv')
raw_csv_data.head(8)
```

```python
# Check for duplicated rows
duplicated_rows = raw_csv_data[raw_csv_data.duplicated()]
duplicated_rows
```

```python
# Drop duplicated rows
raw_csv_data = raw_csv_data.drop_duplicates()
```

```python
df = raw_csv_data.copy()
df.shape
```

```python
# Reset the dataframe index
df.reset_index(drop=True, inplace=True) 
```

```python
# Display all rows and colums
pd.options.display.max_columns = None
pd.options.display.max_rows = None
display(df)
```

```python
df.info()
```

```python
# Drop the 'ID' column
df.drop(['ID'], axis=1, inplace=True)
```

## Working on the "Reasons for absence column"

```python
# Checking for the minimum value
df['Reason for Absence'].min()
```

```python
# Checking for the maximum value
df['Reason for Absence'].max()
```

```python
# Checking for the number of unique values
df['Reason for Absence'].nunique()
```

```python
# Displaying the unique values
df['Reason for Absence'].unique()
```

```python
# Sorting the unique values and displaying on a line
print(sorted(df['Reason for Absence'].unique()), end=' ')
```

### Obtaining dummies

```python
# Create dummies for the 'Reasons' column
reasons_column = pd.get_dummies(df['Reason for Absence'], drop_first=True)
```

```python
reasons_column
```

### Grouping the reasons 

```python
reason_type_1 = reasons_column.loc[:, '1':'14'].max(axis=1) # grouping the first 14 columns and getting the maximum
reason_type_2 = reasons_column.loc[:, '15':'17'].max(axis=1) # grouping the 15th-17th columns and getting the maximum
reason_type_3 = reasons_column.loc[:, '18':'21'].max(axis=1) # grouping the 18th-21st columns and getting the maximum
reason_type_4 = reasons_column.loc[:, '22':].max(axis=1) # grouping the 22nd-28th columns and getting the maximum
```

### Concatenating the column values

```python
# Merging the newly created reason columns and dropping the old reason column
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df = df.drop(['Reason for Absence'], axis=1)
df
```

### Assigning column names for the reasons

```python
# Displaying the column names as an array
df.columns.values
```

```python
# Creating a varible with column names and renaming the reason columns
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']

df.columns = column_names
df.head()
```

```python
# Reordering the column names
column_names_reordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df = df[column_names_reordered]
df.head()
```

```python
# Creating a checkpoint
df_reason_mod = df.copy()
```

## Convert the the 'Date' column to timestamp

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date']
```

### Extracting the months and weekdays

```python
# Getting the month values from the 'Date' column
list_months = []

for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)

print(list_months, end=' ')
```

```python
# List comprehension method of outputing the month values
list_month = [date.month for date in df_reason_mod['Date']]
list_month
```

```python
len(list_months)
```

```python
# Creating a new column in the dataframe for the outputed months
df_reason_mod['Month Value'] = list_months
df_reason_mod.head()
```

```python
# Creating a function to output the days of the week from the 'Date' column
def date_to_weekday(day_value):
    return day_value.weekday()
```

```python
# Creating a new column in the dataframe for the weekdays
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

## Dropping the 'Date' column, reordering and creating a checkpoint

```python
df_reason_mod.drop(['Date'], axis=1, inplace=True)
df_reason_mod.columns.values # displaying the column names as an array
```

```python
# Reordering the column names
column_names_reordered_2 = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Month Value','Day of the week','Transportation Expense', 
        'Distance to Work', 'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education','Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
# Outputing the reordered columns
df_reason_mod = df_reason_mod[column_names_reordered_2]
df_reason_mod.head()
```

```python
# Creating a checkpoint
df_reason_mod_2 = df_reason_mod.copy()
```

## Working on the 'Education' column

```python
# Getting the unique values
df_reason_mod_2['Education'].unique()
```

```python
# Counts of each values
df_reason_mod_2['Education'].value_counts()
```

```python
# Mapping the values as '1' signifies high school while the other 3 signifies education level above high schoo
df_reason_mod_2['Education'] = df_reason_mod_2['Education'].map({1:0, 2:1, 3:1, 4:1})
df_reason_mod_2['Education'].unique()
```

```python
# Counts of remapped values
df_reason_mod_2['Education'].value_counts()
```

```python
# Create a final checkpoint
df_preprocessed = df_reason_mod_2.copy()
df_preprocessed.head()
```

```python
## Export the dataset as *.csv file
df_preprocessed.to_csv('C:/Users/HP/Downloads/Absenteeism_preprocessed.csv', index=False)
```
